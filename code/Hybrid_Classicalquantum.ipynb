Oversampling - Generates synthetic examples for the minority class.

## Model Hybrid Classical Quantum Model

## Angle Embedding Technique

In angle embedding, number of qubits used for quantum circuit is equal to number of features available in the dataset. Our dataset contains the 30 features. In Order to avoid the program crash, therfore use the concept of hybrid model.

```python
import pennylane as qml
from silence_tensorflow import silence_tensorflow

silence_tensorflow()
tf.keras.backend.set_floatx('float64')
nqubits  = 4
dev = qml.device("lightning.qubit", wires=nqubits)
@qml.qnode(dev, interface="tf",diff_method="adjoint")

@qml.qnode(dev)
def qnode(inputs, weights):
    qml.AngleEmbedding(inputs, wires=range(nqubits))
    qml.BasicEntanglerLayers(weights, wires=range(nqubits))
    return [qml.expval(qml.PauliZ(wires=i)) for i in range(nqubits)]
```

```python
nlayer = 1
weight_shapes = {"weights": (nlayer, nqubits)}

qlayer = qml.qnn.KerasLayer(qnode, weight_shapes, output_dim=nqubits)
```

```python

# Function to draw the quantum circuit
def draw_circuit():
    inputs = np.random.rand(nqubits)  # Dummy input for the circuit
    weights = np.random.rand(nlayer, nqubits)  # Random weights for 2 layers
    fig,ax = qml.draw_mpl(qnode)(inputs, weights)
    plt.show()


draw_circuit()
```

```python
# Initial model configuration
quantum_model1 = tf.keras.Sequential([
    Dense(units=30, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(units=10, activation='relu'),
    Dense(units=4, activation='relu'),
    qlayer,  # Quantum layer
    Dense(units=1, activation='sigmoid'),
])
```

```python
quantum_model1.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.005)
quantum_model1.compile(opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = quantum_model1.fit(X_train, y_train, epochs=50, batch_size=40, validation_split=0.2)
```

```python
# Plot training & validation loss and accuracy
plt.figure(figsize=(14, 5))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Accuracy plot
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Evaluate the model
loss, accuracy = quantum_model1.evaluate(X_test, y_test)
print(f'Test loss: {loss}, Test accuracy: {accuracy}')
```

```python
# Make predictions on the test set
y_pred_prob = quantum_model1.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)  # Convert probabilities to binary labels
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Print accuracy, precision, recall, and F1-score
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
# Get predicted probabilities for the positive class
y_pred_prob_qnn = quantum_model1.predict(X_test)  # This will return probabilities

# Compute ROC curve and AUC for ANN
fpr_qnn, tpr_qnn, _ = roc_curve(y_test, y_pred_prob_qnn)
auc_qnn = roc_auc_score(y_test, y_pred_prob_qnn)
print("Hybrid Classical Quantum Model AUC:", auc_qnn)

# Plot ROC curve
plt.figure()
plt.plot(fpr_qnn, tpr_qnn, label='QNN ROC curve (AUC = {:.2f})'.format(auc_qnn))
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Hybrid Classical Quantum Model')
plt.legend(loc='lower right')
plt.show()
```

## Amplitude Embedding Technique

```python
import pennylane as qml
from silence_tensorflow import silence_tensorflow

silence_tensorflow()
tf.keras.backend.set_floatx('float64')
nqubits  = 4
dev = qml.device("lightning.qubit", wires=nqubits)
@qml.qnode(dev, interface="tf",diff_method="adjoint")

@qml.qnode(dev)
def qnode(inputs, weights):
    qml.AmplitudeEmbedding(features = [a for a in inputs], wires=range(nqubits), normalize = True, pad_with = 0.)
    qml.BasicEntanglerLayers(weights, wires=range(nqubits))
    return [qml.expval(qml.PauliZ(wires=i)) for i in range(nqubits)]
```

```python
nlayer = 1
weight_shapes = {"weights": (nlayer, nqubits)}

qlayer = qml.qnn.KerasLayer(qnode, weight_shapes, output_dim=nqubits)
```

```python

# Function to draw the quantum circuit
def draw_circuit():
    inputs = np.random.rand(nqubits)  # Dummy input for the circuit
    weights = np.random.rand(nlayer, nqubits)  # Random weights for 2 layers
    fig,ax = qml.draw_mpl(qnode)(inputs, weights)
    plt.show()


draw_circuit()
```

```python
# Initial model configuration
quantum_model2 = tf.keras.Sequential([
    Dense(units=30, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(units=10, activation='relu'),
    qlayer,  # Quantum layer
    Dense(units=1, activation='sigmoid'),
])
```

```python
quantum_model2.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.005)
quantum_model2.compile(opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = quantum_model2.fit(X_train, y_train, epochs=50, batch_size=40, validation_split=0.2)
```

```python
# Plot training & validation loss and accuracy
plt.figure(figsize=(14, 5))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Accuracy plot
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
