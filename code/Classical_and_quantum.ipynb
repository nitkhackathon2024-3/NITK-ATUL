To handle the imbalanceness I am using resampling techniques:
1. Undersampling - random undersampling
2. Oversampling - SMOTE

Undersampling - Reduce the number of instances in the majority class.

Oversampling - Generates synthetic examples for the minority class.

## Angle Embedding Technique

In angle embedding, number of qubits used for quantum circuit is equal to number of features available in the dataset. Our dataset contains the 30 features. In Order to avoid the program crash, we will reduce the number of features till 4 componets using PCA techniques. This techniques helps in maintaing the important features of the dataset.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.shape
```

```python
nqubits  = 4
dev = qml.device("lightning.qubit", wires=nqubits)
@qml.qnode(dev, interface="tf",diff_method="adjoint")

def qnode(inputs, weights):
    qml.AngleEmbedding(inputs, wires=range(nqubits))
    qml.BasicEntanglerLayers(weights=weights, wires=range(nqubits))
    return [qml.expval(qml.PauliZ(wires=[0]))]

```

```python
nlayer = 1
weight_shapes = {"weights": (nlayer, nqubits)}
qlayer = qml.qnn.KerasLayer(qnode, weight_shapes, output_dim=1)
```

```python

# Function to draw the quantum circuit
def draw_circuit():
    inputs = np.random.rand(nqubits)  # Dummy input for the circuit
    weights = np.random.rand(nlayer, nqubits)  # Random weights for 2 layers
    fig,ax = qml.draw_mpl(qnode)(inputs, weights)
    plt.show()


draw_circuit()
```

```python
quantum_model = tf.keras.models.Sequential([qlayer])
opt = tf.keras.optimizers.Adam(learning_rate=0.005)
quantum_model.compile(opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = quantum_model.fit(X_train_pca, y_train, epochs=50, batch_size=40, validation_split=0.2)
```

## Amplitude Embedding
PCA is not required for this embedding method. Using this encoding method, Number of qubits will be equal to logarthmic base 2 of number of features in the dataset. It means we can use 5 qubits because the number of features in dataset is 30.

```python
import pennylane as qml
from silence_tensorflow import silence_tensorflow

silence_tensorflow()
tf.keras.backend.set_floatx('float64')
nqubits  = 5
dev = qml.device("lightning.qubit", wires=nqubits)
@qml.qnode(dev, interface="tf",diff_method="adjoint")

@qml.qnode(dev)
def qnode(inputs, weights):
    qml.AmplitudeEmbedding(features = [a for a in inputs], wires=range(nqubits), normalize = True, pad_with = 0.)
    qml.BasicEntanglerLayers(weights, wires=range(nqubits))
    return [qml.expval(qml.PauliZ(wires=[0]))]
```

```python
nlayer = 1
weight_shapes = {"weights": (nlayer, nqubits)}

qlayer = qml.qnn.KerasLayer(qnode, weight_shapes, output_dim=nqubits)
```

```python

# Function to draw the quantum circuit
def draw_circuit():
    inputs = np.random.rand(nqubits)  # Dummy input for the circuit
    weights = np.random.rand(nlayer, nqubits)  # Random weights for 2 layers
    fig,ax = qml.draw_mpl(qnode)(inputs, weights)
    plt.show()


draw_circuit()
```

```python
quantum_model2 = tf.keras.models.Sequential([qlayer])
opt = tf.keras.optimizers.Adam(learning_rate=0.005)
quantum_model2.compile(opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = quantum_model2.fit(X_train, y_train, epochs=50, batch_size=40, validation_split=0.2)
```
